```python
from collections import defaultdict
import random
import pandas as pd
import seaborn as sns

items = [random.gauss(10, 2) for i in range(1000)]

print(min(items), max(items))

buckets = defaultdict(int)

for item in items:
    bucket_index = int(item)
    buckets[bucket_index] += 1

# Преобразование словаря buckets в DataFrame
df = pd.DataFrame(list(buckets.items()), columns=['bucket_index', 'count'])

# Создание countplot
sns.countplot(x='bucket_index', data=df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


a = 0.0  
b = 1.0  

theoretical_mean = (a + b) / 2
theoretical_variance = (b - a) * 2 / 12

sample_sizes = [2, 5, 10, 50]

num_samples = 1000

sample_means = {n: [] for n in sample_sizes}
for n in sample_sizes:
    for _ in range(num_samples):
        sample = np.random.uniform(a, b, size=n)
        sample_mean = np.mean(sample)
        sample_means[n].append(sample_mean)

fig, axs = plt.subplots(2, 2, figsize=(12, 8))
axs = axs.flatten()

for i, n in enumerate(sample_sizes):
    normal_mean = theoretical_mean
    normal_std = np.sqrt(theoretical_variance / n)
    
    axs[i].hist(sample_means[n], bins=20, density=True, alpha=0.7, edgecolor='black',color='green')
    
    x = np.linspace(a, b, 100)
    pdf = norm.pdf(x, loc=normal_mean, scale=normal_std)
    axs[i].plot(x, pdf, 'r-', linewidth=2, label='Нормальное распределение',color='orange')
    
    axs[i].set_title(f"n = {n}")
    axs[i].set_xlabel("Выборочное среднее")
    axs[i].set_ylabel("Плотность")
    axs[i].legend()

plt.tight_layout()
plt.show()
```
